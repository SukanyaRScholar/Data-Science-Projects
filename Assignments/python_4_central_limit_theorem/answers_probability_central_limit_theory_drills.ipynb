```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import randint
from numpy import mean
```

## Poisson distribution

### Law of large numbers

```
results = []
for num_trials in range(1,1000):
    trials = np.random.poisson(3, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})


df.plot()
```

### Central limit theorem

```
means = [mean(np.random.poisson(3, 1000)) for _ in range(1000)]


plt.hist(means)
plt.show()
```

## Gamma 

### Law of large numbers

```
results = []
for num_trials in range(1,1000):
    trials = np.random.gamma(5, 1, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})


df.plot()



```

### Central limit theorem

```
means = [mean(np.random.gamma(5, 1, 1000)) for _ in range(5000)]


plt.hist(means)
plt.show()
```
